```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
b = pd.read_csv('param_beta_1.csv')
T = pd.read_csv('param_T_1.csv')

plt.figure(figsize=(10, 5))
plt.plot(b, label="beta")
plt.plot(T, label="T")
plt.legend()
plt.title("Parameters")
plt.show()
```
