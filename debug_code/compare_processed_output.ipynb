## Compare Matlab / Python

```python
[res_python,fs1] = sf.read('data/output/output_python.wav');
[res_matlab,fs2] = sf.read('data/output/output_matlab.wav');
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
plt.title("Python")
plt.plot(res_python)

plt.subplot(1, 3, 2)
plt.title("Matlab")
plt.plot(res_matlab)

plt.subplot(1, 3, 3)
plt.title("diff")
plt.plot(res_matlab-res_python)
plt.show()
```

```python
from IPython.display import Audio, display
display(Audio(res_python, rate=fs1))
display(Audio(res_matlab, rate=fs1))
```

# Compare normal / roll

```python
[original,fs1] = sf.read('../data/Music/input_ch1.wav')
[res_kalman_speech,fs1] = sf.read('../data/Music/input_ch1_shat.wav')

[res_speech,fs2] = sf.read('../data/Output/s_origin_music_1.wav')
[res_roll_speech,fs2] = sf.read('../data/Output/s_origin_music_2.wav')


plt.figure(figsize=(25, 5))
plt.subplot(1, 4, 1)
plt.title("Original speech")
plt.plot(original)

plt.subplot(1, 4, 2)
plt.title("Kalman speech output")
plt.plot(res_kalman_speech, label="Kalman")
plt.plot(res_speech, label="Python")
plt.legend()


plt.subplot(1, 4, 3)
plt.title("Python speech & Roll output")
plt.plot(res_speech, label="Speech")
plt.plot(res_roll_speech, label="Rolled speech")
plt.legend()

plt.subplot(1, 4, 4)
# plt.title("diff")
# plt.plot(res_kalman_original[:len(res_python_speech)] - res_python_speech)
plt.title("Python Roll output")
plt.plot(res_roll_speech)

#  
plt.legend()
plt.show()
```

```python
[res_kalman_noise,fs1] = sf.read('../data/Music/input_ch1_dhat.wav')
[res_roll_noise,fs2] = sf.read('../data/Output/n_origin_music_2.wav')

plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
plt.title("kalman")
plt.plot(res_kalman_noise)

plt.subplot(1, 3, 2)
plt.title("Python")
plt.plot(res_roll_noise)

plt.subplot(1, 3, 3)
plt.title("diff")
plt.plot(res_kalman_noise[:len(res_roll_noise)] - res_roll_noise)
plt.show()
```
